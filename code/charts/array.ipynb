```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'SimHei'
```

```python
# 单向链表查询速度曲线图
import pandas as pd

# 1. 读取 CSV 文件
df = pd.read_csv('array_insert.csv', sep='\t', header=None, names=['Size', 'Spend'])  # 替换为你的文件路径

# 3. 绘制折线图
plt.figure(figsize=(10, 6))
plt.plot(df['Size'], df['Spend'], color='red', linewidth=2)
plt.title('数组插入速度曲线')
plt.xlabel('元素个数')
plt.ylabel('插入耗时')
plt.grid(True)
plt.tight_layout()
plt.show()
```
